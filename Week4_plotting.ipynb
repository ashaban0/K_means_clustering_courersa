```python
from pandas import read_csv
import numpy as np
```

```python
df = read_csv('happyscore_income.csv')
df.sort_values('avg_income', inplace=True)

avg_income = np.array(df.avg_income)
happyScore = np.array(df.happyScore)
country = np.array(df.country)
df
```

```python
import matplotlib.pyplot as plt

plt.close()
fig = plt.figure(1, figsize=(6,4), dpi=300)
ax1 = fig.add_subplot(111)
ax1.plot(avg_income, happyScore, 'bo')
ax1.text(avg_income[0], happyScore[0], country[0])
ax1.text(avg_income[-1], happyScore[-1], country[-1])
ax1.set_xlabel('avgerage income')
ax1.set_ylabel('happy Score')

fig.tight_layout()
plt.show()
```

**Why you chose to plot those particular columns of data?**  


I chose to plot the "happy Score" versus average income because I think they are correlated. As you increase the average income, the "happy Score" increase. I wanted to visualize that for the whole dataset.

**how you sorted or filtered the data?**

I sorted the data according to the average income to make the countries in increasing order from the lowest average income to the highest average income.

**what makes the labelled points interesting?**

The labelled points represent the countries with the lowest income "Burundi" and the one with the highest income "Luxembourg".
