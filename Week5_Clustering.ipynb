```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.cluster import KMeans
import numpy as np
```

### Load the dataset

```python
df = pd.read_csv('Banknote-authentication-dataset.csv')
v1 = np.array(df['V1'])
v2 = np.array(df['V2'])


data = np.column_stack((v1, v2))
print(data)
```

#### 1. run K-means on the given dataset.
#### 2. visualize the results

```python
km_results = KMeans(n_clusters=2).fit(data)

clusters = km_results.cluster_centers_


plt.close()
fig = plt.figure(1, figsize=(6,4), dpi=300)
ax1 = fig.add_subplot(111)
ax1.scatter(v1, v2, s=5)
ax1.scatter(clusters[:,0], clusters[:,1], s=100)

ax1.set_xlabel("V1 variance")
ax1.set_ylabel("V2 Skewness")
fig.tight_layout()
plt.show()
```

#### 3) re-run K-means several times and visualize the results

```python
km_results = KMeans(n_clusters=2).fit(data)

clusters = km_results.cluster_centers_


plt.close()
fig = plt.figure(1, figsize=(6,4), dpi=300)
ax1 = fig.add_subplot(111)
ax1.scatter(v1, v2, s=5, label="Data")
ax1.scatter(clusters[:,0], clusters[:,1], s=100, label="Clusters Centers")

ax1.legend(loc=0)
ax1.set_xlabel("V1 variance")
ax1.set_ylabel("V2 Skewness")

fig.tight_layout()
plt.show()
fig.savefig('k_means_clustering_example.png', bbox_inches='tight', dpi=300)
```

#### 4) compare the results: is the K-means algorithm stable?


The results seems to be consistent. This means that the K-means algorithm is stable.
