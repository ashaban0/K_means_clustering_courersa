## Code and analysis

### Loading the important libraries:

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
```

### 1) Load the CSV file

```python
df = pd.read_csv('BankNote_Authentication.csv')
var = np.array(df['variance'])
skew = np.array(df['skewness'])
curt = np.array(df['curtosis'])
entr = np.array(df['entropy'])
clas = np.array(df['class'])

df1 = df[ df['class'] > 0 ]
df0 = df[ df['class'] < 1]

v_list = [var, skew, curt, entr]
v_names = ["Variance", "Skewness", "Curtosis", "Entropy"]
df
```

### 2) Calculate statistical measures, e.g. mean and standard devaition

```python
for i in range(len(v_list)):
    m = np.mean(v_list[i])
    std = np.std(v_list[i])
    print(v_names[i]," Mean: ", m, ", standard deviation: ", std)
```

### 3) Visualize your data as you consider fit

```python
mark = 10
#### Skewness versus variance of wavelet
plt.close()
fig = plt.figure(1, figsize=(6,6), dpi=300)
ax1 = fig.add_subplot(111)
ax1.scatter(df0.variance, df0.skewness, s=10, color='b', label='0')
ax1.scatter(df1.variance, df1.skewness, s=10, color='orange', label='1')

ax1.plot(np.mean(df0.variance), np.mean(df0.skewness), 'kx', markersize=mark)
ax1.plot(np.mean(df1.variance), np.mean(df1.skewness), 's', c='k', markersize=mark)

ax1.set_xlabel("V1 Variance of Wavelet")
ax1.set_ylabel("V2 Skewness of Wavelet")
ax1.legend(loc=0)


fig.tight_layout()
plt.show()
```
