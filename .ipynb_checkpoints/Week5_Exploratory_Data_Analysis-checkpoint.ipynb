## Code and analysis

### Loading the important libraries:

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.patches import Ellipse
```

### 1) Load the CSV file

```python
df = pd.read_csv('BankNote_Authentication.csv')
var = np.array(df['variance'])
skew = np.array(df['skewness'])
curt = np.array(df['curtosis'])
entr = np.array(df['entropy'])
clas = np.array(df['class'])

df1 = df[ df['class'] > 0 ]
df0 = df[ df['class'] < 1]

v_list = [var, skew, curt, entr]
v_names = ["Variance", "Skewness", "Curtosis", "Entropy"]
df
```

### 2) Calculate statistical measures, e.g. mean and standard devaition

```python
for i in range(len(v_list)):
    m = np.mean(v_list[i])
    std = np.std(v_list[i])
    print(v_names[i]," Mean: ", m, ", standard deviation: ", std)
```

```python
mean_v_0 = np.mean(df0.variance) 
mean_v_1 = np.mean(df1.variance)

mean_sk_0 = np.mean(df0.skewness)
mean_sk_1 = np.mean(df1.skewness)

std_v_0 = np.std(df0.variance)
std_v_1 = np.std(df1.variance)

std_sk_0 = np.std(df0.skewness)
std_sk_1 = np.std(df1.skewness)
```

### 3) Visualize your data as you consider fit

```python
mark = 10
#### Skewness versus variance of wavelet
plt.close()
fig = plt.figure(1, figsize=(6,6), dpi=300)
ax1 = fig.add_subplot(111)
ax1.scatter(df0.variance, df0.skewness, s=10, color='b', label='0')
ax1.scatter(df1.variance, df1.skewness, s=10, color='orange', label='1')

ax1.plot( mean_v_0, mean_sk_0, 'kx', markersize=mark)
ax1.plot( mean_v_1, mean_sk_1, 's', c='k', markersize=mark)

ax1.text(mean_v_0+0.3, mean_sk_0, "Mean 0", fontsize='15', color='m')
ax1.text(mean_v_1+0.3, mean_sk_1, "Mean 1", fontsize='15', color='m')


e0 = Ellipse( (mean_v_0, mean_sk_0), width=3*std_v_0, height=3*std_sk_0, alpha=0.2, color='b' )
e1 = Ellipse( (mean_v_1, mean_sk_1), width=3*std_v_1, height=3*std_sk_1, alpha=0.2, color='orange' )

ax1.set_xlabel("V1 Variance of Wavelet")
ax1.set_ylabel("V2 Skewness of Wavelet")

ax1.add_patch(e0)
ax1.add_patch(e1)
ax1.legend(loc=0)


fig.tight_layout()
plt.show()

fig.savefig('Skewness_Vs_Variance_Wavelet_BankNote.png', bbox_inches='tight', dpi=300)
```

#### Figure caption
Skewness of the wavelet on the y-axis versus the variance of the wavelet on the x-axis. Blue datapoints represent the counterfeit banknotes. Orange datapoints represent the real banknotes. The blue ellipse is centered on the mean of the counterfeit with width and height equal to 2 standard deviations of the variance and skewness of the counterfeit. The orange ellipse is centered on the mean of the real banknotes with width and height equal to 2 standard deviations of the variance and skewness of the real banknotes.
